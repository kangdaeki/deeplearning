```python
import numpy as np
def step(x):
	result = x > 0.000001		# True 또는 False. 원래는 0보다 큰 경우인데 부동소수점 오차가 있으므로 이렇게 함
	return result.astype(np.int64)	# 정수로 반환
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10.0, 10.0, 0.1)
y = step(x)
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
	return 1.0 / (1.0 + np.exp(-x)) 

x = np.arange(-10.0, 10.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
  
x = np.linspace(-np.pi, np.pi, 60) 
y = np.tanh(x) 
plt.plot(x, y) 
plt.show() 
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

def relu(x):
	return np.maximum(x, 0)
  
x = np.arange(-10.0, 10.0, 0.1)
y = relu(x)
plt.plot(x, y) 
plt.show() 
```
